### Functions

```python
# Declaring a linear function in Python

def f(x):
    return 2 * x + 1

x_values = [0, 1, 2, 3]

for x in x_values:
    y = f(x)
    print(y)
```

```python
# Charting a linear function using sympy

x = sympy.symbols("x")
f = (2 * x) + 1
sympy.plot(f)
```

```python
# Charting an exponential function

x = sympy.symbols("x")
f = (x**2) + 1
sympy.plot(f)
```

```python
# Declaring a function with two independent variables

x, y = sympy.symbols("x y")
f = (2 * x) + (3 * y)
sympy.plotting.plot3d(f)
```

### Summations

```python
# Performing a summation

summation = sum(2 * i for i in range(1, 6))
print(summation)
```

```python
# Summation of elements

x = [1, 4, 6, 2]
n = len(x)

summation = sum(10 * x[i] for i in range(0, n))
print(summation)
```

```python
# Summation in sympy

i, n = sympy.symbols("i n")
summation = sympy.Sum(2 * i, (i, 1, n))
summation_up_to_five = summation.subs(n, 5)
print(summation_up_to_five.doit())
```

### Exponents

```python
# Simplify expression with sympy

x = sympy.symbols("x")
expr = (x ** 2) / (x ** 5)
print(expr)
```

### Logarithms

```python
# Using the log function

x = math.log(8, 2)
print(x)
```

```python
# Calculating compound interest

initial_investment = 100
rate = 0.20
time_span = 2.0
periods = 12

balance = initial_investment * (1 + (rate/periods)) ** (periods * time_span)
print(balance)
```
