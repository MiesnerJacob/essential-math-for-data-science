# Chapter 02: Probability

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import beta
```

## Concepts

### Joint Probabilities

```python
# Joint probability (AND) or independent events

prob_event_a = 0.5
prob_event_b = 0.25

prob_event_a_and_b = prob_event_a * prob_event_b
print(prob_event_a_and_b)
```

### Union Probabilities

```python
# Union Probability (OR) of mutually exclusive events

prob_event_a = 0.5
prob_event_b = 0.25

prob_event_a_or_b = prob_event_a + prob_event_b
print(prob_event_a_or_b)
```

```python
# Union probability (OR) of non mutually exclusive events

prob_event_a = 0.5
prob_event_b = 0.25

prob_event_a_or_b = prob_event_a + prob_event_b - (prob_event_a * prob_event_b)
print(prob_event_a_and_b)
```

### Conditional Probability

```python
# Calculate conditional probability with Bayes Theorem

prob_event_a = 0.65
prob_event_b = 0.05
prob_event_b_given_a = 0.85

prob_event_a_given_b = (prob_event_b_given_a * prob_event_b) / prob_event_a
print(prob_event_a_given_b)
```

### Joint and Union Conditional Probabilities

```python
# Joint Conditional Probability

prob_event_a = 0.005
prob_event_b_given_a = 0.85

prob_event_a_and_b = prob_event_b_given_a * prob_event_a
print(prob_event_a_and_b)
```

```python
# Union Conditional Probability

prob_event_a = 0.005
prob_event_b = 0.5
prob_event_b_given_a = 0.85

prob_event_a_or_b = prob_event_a + prob_event_b - (prob_event_b_given_a * prob_event_b)
print(prob_event_a_or_b)
```

### Binomial Distribution

```python
# Parameters
n = 10
p = 0.9

# Calculate the binomial distribution
x = np.arange(0, n+1)
probabilities = binom.pmf(x, n, p)

for index, value in enumerate(probabilities):
    print("{0} - {1}".format(index, value))

# Plot
plt.figure(figsize=(5, 4))
plt.bar(x, probabilities, color='blue', alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.grid(True)
plt.show()
```

### Beta Distribution

```python
# Parameters for the beta distribution
alpha = 8
beta_param = 2

# Calculate the prbability of the true value being less than 0.9
cdf_values = beta.cdf(.9, alpha, beta_param)
print("probability of true value being <0.9:",cdf_values)

# Calculate the beta distribution
x = np.linspace(0, 1, 100)
probabilities = beta.pdf(x, alpha, beta_param)

# Plot
plt.figure(figsize=(5, 4))
plt.plot(x, probabilities, color='blue', alpha=0.7, label='Beta PDF')
plt.fill_between(x, 0, probabilities, where=(x < 0.9), color='blue', alpha=0.3, label='Area where x < 0.9')
plt.axvline(x=0.9, color='red', linestyle='--', label='x = 0.9')
plt.xlabel('x')
plt.ylabel('Density')
plt.title(f'Beta Distribution (alpha={alpha}, beta={beta_param})')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Parameters for the beta distribution
alpha = 8
beta_param = 2

# Calculate the prbability of the true value being greater than 0.9
cdf_values = 1 - beta.cdf(.9, alpha, beta_param)
print("Probability of true value being >0.9:",cdf_values)

# Calculate the beta distribution
x = np.linspace(0, 1, 100)
probabilities = beta.pdf(x, alpha, beta_param)

# Plot
plt.figure(figsize=(5, 4))
plt.plot(x, probabilities, color='blue', alpha=0.7, label='Beta PDF')
plt.fill_between(x, 0, probabilities, where=(x > 0.9), color='blue', alpha=0.3, label='Area where x < 0.9')
plt.axvline(x=0.9, color='red', linestyle='--', label='x = 0.9')
plt.xlabel('x')
plt.ylabel('Density')
plt.title(f'Beta Distribution (alpha={alpha}, beta={beta_param})')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Parameters for the beta distribution
alpha = 8
beta_param = 2

# Calculate the prbability of the true value between 0.8 and 0.9
cdf_values = beta.cdf(.9, alpha, beta_param) - beta.cdf(.8, alpha, beta_param)
print("Probability of true value being between 0.8-0.9:",cdf_values)

# Calculate the beta distribution
x = np.linspace(0, 1, 100)
probabilities = beta.pdf(x, alpha, beta_param)

# Plot
plt.figure(figsize=(5, 4))
plt.plot(x, probabilities, color='blue', alpha=0.7, label='Beta PDF')
plt.fill_between(x, 0, probabilities, where=(x >= 0.8) & (x <= 0.9), color='blue', alpha=0.3, label='Area between 0.8 and 0.9')
plt.axvline(x=0.8, color='green', linestyle='--', label='x = 0.8')
plt.axvline(x=0.9, color='red', linestyle='--', label='x = 0.9')
plt.xlabel('x')
plt.ylabel('Density')
plt.title(f'Beta Distribution (alpha={alpha}, beta={beta_param})')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Exercise 2-4 (Binomial Distribution: Probability of sample given true value)

passengers_las_vegas_to_dallas = 137
prob_passenger_shows_up = 0.4

# Calculate the binomial distribution
x = np.arange(0, passengers_las_vegas_to_dallas+1)
probabilities = binom.pmf(x, passengers_las_vegas_to_dallas, prob_passenger_shows_up)

prob_50_or_more_passengers_no_show = sum(probabilities[50:])
print("Probability 50 or more people no show:", prob_50_or_more_passengers_no_show)

# Plot
plt.figure(figsize=(5, 4))
plt.bar(x, probabilities, color='blue', alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={passengers_las_vegas_to_dallas}, p={ prob_passenger_shows_up})')
plt.grid(True)
plt.show()
```

```python
# Exercise 2-5 (Beta Distributuion: Probability of true value given sample)

coin_flips = 19
num_heads = 15
num_tails = 4

# Calculate the prbability of the true value between 0.45 and 0.55 (roughly fair coin)
cdf_values = beta.cdf(.6, num_heads, num_tails) - beta.cdf(.4, num_heads, num_tails)
print("Probability of true value being between 0.4-0.6:",cdf_values)

# Calculate the beta distribution
x = np.linspace(0, 1, 100)
probabilities = beta.pdf(x, num_heads, num_tails)

# Plot
plt.figure(figsize=(5, 4))
plt.plot(x, probabilities, color='blue', alpha=0.7, label='Beta PDF')
plt.fill_between(x, 0, probabilities, where=(x >= 0.4) & (x <= 0.6), color='blue', alpha=0.3, label='Area between 0.4 and 0.6')
plt.axvline(x=0.4, color='green', linestyle='--', label='x = 0.4')
plt.axvline(x=0.6, color='red', linestyle='--', label='x = 0.6')
plt.xlabel('x')
plt.ylabel('Density')
plt.title(f'Beta Distribution (alpha={num_heads}, beta={num_tails}')
plt.legend()
plt.grid(True)
plt.show()

print("The probability of this being a fair coin is low, given the information from our beta distribution. Although, more flips would give us grater confidence.")
```
